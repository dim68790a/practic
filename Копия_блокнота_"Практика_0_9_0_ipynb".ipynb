# Задание 0

Создайте функцию по нахождению уникальных элементов из двух списков



```
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
```



````python
def sort(a, b):
    z=''
    for i in range(len(a)):
        if not(a[i] in b):
            z += str(a[i]) + ' '
        if not(b[i] in a):
            z += str(b[i]) + ' '
    return z

a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
print(sort(a, b))
````

# Задание 1

Симулятор виртуального питомца

Цель: создать виртуальный симулятор домашних животных, в котором пользователи смогут заводить питомцев и ухаживать за ними.

Требования:

- Функция для усыновления питомца (имя, тип, возраст).
- Функция для того, чтобы покормить питомца, поиграть с ним или уложить его спать.
- Функция для отображения состояния питомца (голод, радость, энергия).

````python
def pet():
    animal={}
    animal['name']=input('Давай создадим питомца: придумай ему имя ')
    animal['pet']=input('Кто это будет(собака, кот, дракон ...): ')
    animal['age']=input('Сколько ему лет: ')
    return animal

def play():
    global fun, eats, energy
    fun += 15
    eats -= 10
    energy -= 10

def eat():
    global fun, eats, energy
    fun -= 5
    eats += 15
    energy += 5

def slip():
    global fun, eats, energy
    fun -= 5
    eats -= 5
    energy += 15
def stat():
    global fun, eats, energy
    if fun < 35:
        if fun <= 0:
            print('Game over: ему было слишком скучно')
            return 1
        else:
            print('Я хочу играть')
    if eats < 35:
        if eats <= 0:
            print('Game over: он был голоден')
            return 1
        else:
            print('Я хочу кушать')
    if energy < 35:
        if energy <= 0:
            print('Game over: переутомился')
            return 1
        else:
            print('Я хочу спать')

fun = 50
eats = 50
energy = 50
animal = pet()
while True:
    if stat() == 1:
        break
    print(f'Что будем делать вместе с {animal["name"]}:')
    print('1. играть')
    print('2. кушать')
    print('3. спать')
    print('4. посмотреть данные')
    question = int(input())
    if question == 1:
        play()
    elif question == 2:
        eat()
    elif question == 3:
        slip()
    else:
        print(f'голод: {eats}, радость: {fun}, энергия: {energy}')
````

# Задание 2

Рыцарь и дракон

Цель: создать небольшую игру, в которой вам необходимо играть за рыцаря и сразиться с драконом

Требования:

- Создание персонажа (имя, информация о доспехах, оружии, урон, здоровье)
- Управление персонажем и мини сюжет
- Создание дракона (Имя, информация о здоровье и уроне)
- Боевая система (нанесение и получение урона, урон должен быть случайным в заданном диапазоне)
- Реализовать бой между драконом и рыцарем

````python
import random as rm

def box():
    global hero, hp
    lut = rm.randint(0, 100)
    dm = rm.randint(10, 30)
    armor = rm.randint(5, 15)
    if 0<=lut<=30:
        print(f"там лежал меч с урон {dm}")
        question = int(input("Берём?: 1. да 2. нет "))
        if question == 1:
            hero['оружие']=dm
    elif 31<=lut<=61:
        print(f"там лежали доспехи с защитой {armor}")
        question = int(input("Берём?: 1. да 2. нет "))
        if question == 1:
            hero['защита']=armor
    elif 62<=lut<=90:
        print(f"там лежали меч и доспехи с уроном {dm} и защитой {armor}")
        question = int(input("Берём?: 1. только меч 2. только защиту 3. всё "
        "4. нет "))
        if question == 1:
            hero['оружие']=dm
        elif question == 2:
            hero['защита'] = armor
        elif question == 3:
            hero['защита'] = armor
            hero['оружие'] = dm
    else:
        print(f"там была ловушка вы потеряли {dm}")
        hero["здоровье"] -= dm


def pers():
    global hp, damage
    hero = {"здоровье": hp, 'урон': damage, "защита": 0, "оружие": 0}
    hero['name'] = input('Как назовём героя? ')
    print(f'Вы {hero} у вас {hp} единеиц здоровья и {damage} урона')
    print('доспехов и оружия нету')
    return hero

def dragon():
    enemy = {"Здоровье": rm.randint(100, 250), "урон": rm.randint(35, 60)}
    return enemy

def fight():
    global drag, hero
    print("бой начинается")
    while hero["здоровье"] > 0 and drag['Здоровье'] > 0:
        do_drag=rm.randint(1, 3)
        dam = hero['урон']+hero['оружие']-10 + rm.randint(0, 20)
        dam_dr = drag['урон'] - 10 + rm.randint(0, 20)
        do_hero = int(input("что будем делать 1. атаковать 2.уклоняться "))
        if do_hero == 1:
            if do_drag == 1:
                hero["здоровье"] = hero["здоровье"] - abs(hero["защита"]-dam_dr)
                print('Дракон тоже атаковал но он сильнее ВЫ ПОЛУЧИЛИ УРОН')
            elif do_drag == 2:
                drag['Здоровье'] = drag['Здоровье'] - (dam//2)
                print('дракон выставил своё крыло вы нанесли ', dam//2)
            else:
                drag['Здоровье'] = drag['Здоровье'] - dam
                print('Дракон отдыхал вы нанесли', dam)
        else:
            if do_drag == 1:
                print('Дракон атаковал но вы увернулись')
            elif do_drag == 2:
                hero['здоровье'] = hero['здоровье'] - abs(hero["защита"]
                                                          -(dam_dr//2))
                print('дракон дыхнул огнем но вы увернулись вас только зацепило,'
                ' вам нанесли',
                      abs(hero["защита"]-(dam_dr//2)))
            else:
                print('Дракон отдыхал')
    if hero["здоровье"] > 0:
        return 1
    else:
        return 0




hp = 40
damage = 10
hero = pers()
print('вы сидите в темнице, перед вами сундук')
question = int(input('открыть?: 1. да 2. нет' ))
if question == 1:
    box()
num = rm.randint(1, 5)
print(f'впереди {num} драконов')
while num > 0 and hero["здоровье"] > 0:
    drag = dragon()
    print(f'Впреди дракон с {drag["Здоровье"]} здоровье и {drag["урон"]} уроном')
    result = fight()
    if result == 1:
        print('ВЫ ПОБЕДИЛИ')
        question = int(input('вы нашли сундук, открыть?: 1. да 2. нет '))
        if question == 1:
            box()
        num -= 1
    else:
        print('вы погибли')
        break
````

# Задание 3


Цель - создать менеджера команды Pokémon, который позволит пользователям:

- Добавлять покемонов в свою команду. (если такого покемона еще нет в команде)
- Удалять покемонов из их команды.
- Просматривать подробную информацию обо всех покемонах в команде.
- Находить покемона по имени.
- Устраивать тренировочный бой между двумя покемонами

Для данной задачи используйте: https://pokeapi.co/


````python
import requests


def get_pokemon_data(name):
    url = f"https://pokeapi.co/api/v2/pokemon/{name}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        return None

def add_pokemon(team):
    name = input("Введите имя покемона: ").lower()
    data = get_pokemon_data(name)
    if data:
        if name not in team:
            team[name] = data
            print(f"Покемон {name} добавлен в команду!")
        else:
            print(f"Покемон {name} уже есть в команде.")
    else:
        print(f"Покемон {name} не найден.")


def remove_pokemon(team):
    name = input("Введите имя покемона для удаления: ").lower()
    if name in team:
        del team[name]
        print(f"Покемон {name} удален из команды.")
    else:
        print(f"Покемон {name} не найден в команде.")


def view_pokemon(team):
    if not team:
        print("Команда пуста.")
        return
    for name, data in team.items():
        print(f"Имя: {name}\nТип: {', '.join([t['type']['name'] for t in data['types']])}\n"
              f"Атака: {data['stats'][1]['base_stat']}\nЗащита: "
              f"{data['stats'][2]['base_stat']}\n"
              f"HP: {data['stats'][0]['base_stat']}")


def find_pokemon():
    name = input("Введите имя покемона для поиска: ").lower()
    data = get_pokemon_data(name)
    if data != None:
        print(f"Имя: {name}\nТип: {', '.join([t['type']['name'] for t in data['types']])}\n"
              f"Атака: {data['stats'][1]['base_stat']}\nЗащита: "
              f"{data['stats'][2]['base_stat']}\n"
              f"HP: {data['stats'][0]['base_stat']}")
    else:
        print(f"Покемон {name} не найден в команде.")


def training_battle(team):
    if len(team) < 2:
        print("Для тренировочного боя нужно минимум два покемона.")
        return
    print("Список покемонов:")
    for i, name in enumerate(team):
        print(f"{i+1}. {name}")
    try:
        pokemon1_index = int(input("Выберите первого покемона: ")) - 1
        pokemon2_index = int(input("Выберите второго покемона: ")) - 1
        pokemon1_name = list(team.keys())[pokemon1_index]
        pokemon2_name = list(team.keys())[pokemon2_index]
        pokemon1_data = team[pokemon1_name]
        pokemon2_data = team[pokemon2_name]

        pokemon1_hp = pokemon1_data['stats'][0]['base_stat']
        pokemon1_attack = pokemon1_data['stats'][1]['base_stat']
        pokemon1_defense = pokemon1_data['stats'][2]['base_stat']

        pokemon2_hp = pokemon2_data['stats'][0]['base_stat']
        pokemon2_attack = pokemon2_data['stats'][1]['base_stat']
        pokemon2_defense = pokemon2_data['stats'][2]['base_stat']

        pokemon1_score = (pokemon1_hp + pokemon1_defense) / pokemon1_attack
        pokemon2_score = (pokemon2_hp + pokemon2_defense) / pokemon2_attack

        if pokemon1_score > pokemon2_score:
            print(f"{pokemon1_name} победил!")
        elif pokemon2_score > pokemon1_score:
            print(f"{pokemon2_name} победил!")
        else:
            print("Ничья!")
        print("Бой закончен!")
    except (ValueError, IndexError):
        print("Некорректный ввод.")


team = {}
while True:
    print("Меню:")
    print("1. Добавить покемона")
    print("2. Удалить покемона")
    print("3. Просмотреть информацию о покемонах")
    print("4. Найти покемона")
    print("5. Тренировочный бой")
    print("6. Выход")

    choice = input("Выберите действие: ")

    if choice == '1':
        add_pokemon(team)
    elif choice == '2':
        remove_pokemon(team)
    elif choice == '3':
        view_pokemon(team)
    elif choice == '4':
        find_pokemon()
    elif choice == '5':
        training_battle(team)
    elif choice == '6':
        break
    else:
        print("Некорректный выбор.")
````
